# Housing Trends in Philadelphia, PA

*Glossary*

**Home values = Zillow Home Value Index (ZHVI)**, a measure of the typical home value and market changes across a given region and housing type, reflecting the typical value for homes in the 35th to 65th percentile range.

**Rent prices = Zillow Observed Rent Index (ZORI)**, a measure of the typical observed market rate rent across a given region, computed as the mean of listed rents that fall into the 40th to 60th percentile range for all homes and apartments in a given region, weighted to reflect the rental housing stock (i.e., so it is not only reflective of currently listed rentals). 

Learn more: https://www.zillow.com/research/data/
